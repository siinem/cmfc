```python
# --------------------------------------------------------
# This exercise is based on the code of Alfio Ferrara
# Repository: https://github.com/umilISLab/midh/tree/main
# --------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt

from PIL import Image

cat_matrix = np.array([
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255],
    [255, 0, 0, 255, 255, 255, 255, 0, 0, 255],
    [255, 0, 0, 255, 255, 255, 255, 0, 0, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255],
    [255, 255, 255, 255, 0, 0, 255, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255],
    [255, 255, 0, 255, 255, 255, 255, 0, 255, 255],
    [255, 255, 0, 0, 0, 0, 0, 0, 255, 255],
    [255, 255, 255, 190, 0, 0, 190, 255, 255, 255],
    [255, 255, 255, 255, 255, 255, 255, 255, 255, 255]
])


cat_matrix = 255-cat_matrix
plt.imshow(cat_matrix, cmap='gray', vmin=0, vmax=255)
plt.title("Smile face image")
plt.axis('off')  # Hide the axes
plt.show()
```

```python
# Create a full RGB image from the cat_matrix
# Each pixel is [R, G, B]
# For simplicity, we'll assume a simple mapping for RGB
red_channel = cat_matrix   # Red values
green_channel = np.zeros_like(cat_matrix)    # Green values (all zeros)
blue_channel = np.zeros_like(cat_matrix)    # Blue values

print(red_channel)
print(green_channel)
print(blue_channel)
```

```python

# Stack the channels to create an RGB image

red_channel = cat_matrix   # Red values
green_channel = np.zeros_like(cat_matrix)    # Green values (all zeros)
blue_channel = np.zeros_like(cat_matrix)    # Blue values
rgb_coloredRed = np.stack((red_channel, green_channel, blue_channel), axis=-1)

red_channel =  np.zeros_like(cat_matrix)    # Red values
green_channel = cat_matrix   # Green values (all zeros)
blue_channel = np.zeros_like(cat_matrix)    # Blue values
rgb_coloredGreen = np.stack((red_channel, green_channel, blue_channel), axis=-1)


red_channel = np.zeros_like(cat_matrix)   # Red values
green_channel = np.zeros_like(cat_matrix)    # Green values (all zeros)
blue_channel = cat_matrix     # Blue values
rgb_coloredBlue = np.stack((red_channel, green_channel, blue_channel), axis=-1)

# Display the RGB image
#plt.figure(figsize=(10, 10))
#plt.imshow(rgb_image)
#plt.title("Combined RGB Image")
#plt.axis('off')  # Hide the axes
#plt.show()

# Visualize each channel separately
plt.figure(figsize=(15, 5))

# Red channel
plt.subplot(1, 3, 1)
plt.imshow(rgb_coloredRed, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')

# Green channel
plt.subplot(1, 3, 2)
plt.imshow(rgb_coloredGreen, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')

# Blue channel
plt.subplot(1, 3, 3)
plt.imshow(rgb_coloredBlue, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

plt.show()
```
